## LOADING THE DATASET USING PANDAS

```python
import numpy as np
import pandas as pd
df=pd.read_csv('MYNTRA DATA.csv')
df
```

## CLEANING THE DATASET

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#deleting columns that are not neccessary for the analysis 
df=df.drop(['sizes','product_link','img_link'],axis=1)
```

```python
#deleting rows containing none value in rating_count column
df=df.dropna(subset=['rating_count'])
df=df.reset_index(drop=True)
df.info()
```

```python
#replacing discount_percent column's null values by discount pecentage calculated by the percentage formula  
fill=(df['discounted_price']/df['marked_price'])*100
df['discount_percent']=df['discount_percent'].fillna(fill)
df['discount_percent']=df['discount_percent'].round(0)
df
```

```python
df.info()
```

```python
# rating_count column contain string value, to find it we have following code
df[~df['rating_count']
      .astype(str)
      .str.replace('.', '', 1)
      .str.isnumeric()]
```

```python
# only one row contain a symbol '\', we can delete the entire row
df=df.drop(index=216)
```

```python
# converting rating_count column data type from object to integer
df['rating_count']=df['rating_count'].astype(int)
```

```python
df.info()
```

## TRANSFERING THE CLEANED DATA TO MYSQL

```python
import pandas as pd
from sqlalchemy import create_engine
import urllib.parse

# Database credentials
username = "root"
password = "@Nayan2897"  # contains special character '@'
host = "localhost"
port = 3306
database = "myntra1"

# URL-encode the password to handle special characters
encoded_password = urllib.parse.quote_plus(password)

# Create SQLAlchemy engine
engine = create_engine(
    f"mysql+pymysql://{username}:{encoded_password}@{host}:{port}/{database}"
)


# Write DataFrame to MySQL
table_name = "data"
try:
    df.to_sql(table_name, engine, if_exists="replace", index=False)
    print(f"Data written to table '{table_name}' successfully.")
except Exception as e:
    print("Error writing to database:", e)

# Read back sample data
try:
    result_df = pd.read_sql("SELECT * FROM customer LIMIT 5;", engine)
    print(result_df)
except Exception as e:
    print("Error reading from database:", e)
```
